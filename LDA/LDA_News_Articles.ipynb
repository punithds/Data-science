```python
# importing necessary libraries
import warnings
warnings.filterwarnings("ignore")
import pandas as pd 
pd.set_option("display.max_columns",999)
import os

# libraries for visualizing the topics
import pyLDAvis 
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# loading the dataset
df=pd.read_csv("NewsArticles.csv",encoding= 'unicode_escape')
df.head()
```

```python
# we will consider  text column for extracting the topics
df=df["text"]
df
```

```python
# Defining tf features for LDA..."
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                max_features=1000,
                                stop_words='english')
```

```python
tf = tf_vectorizer.fit_transform(df.values.astype('U'))
```

```python
# Defining LDA parameteres
lda = LatentDirichletAllocation(n_components=10, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
lda.fit(tf)
```

```python
tf_feature_names=tf_vectorizer.get_feature_names()
print(tf_feature_names)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += " ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words:-1]])
        print(message)
```

```python
# printing the Top 10 Topics
print_top_words(lda,tf_feature_names,10)
```

```python
# Visualize Topics using pylda
number_topics=10
count_data=tf
count_vectorizer=tf_vectorizer
pyLDAvis.sklearn.prepare(lda, count_data, count_vectorizer)
```
